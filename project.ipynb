```python
# Calculate the RSI for the entire dataset
def calculate_rsi(data, period=14):
    close_delta = data['Close'].diff()
    gain = (close_delta.where(close_delta > 0, 0)).rolling(window=period).mean()
    loss = (-close_delta.where(close_delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

combined_data['RSI'] = calculate_rsi(combined_data)
```

```python
#quarter wise price trend
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming combined_data contains the Date and Close columns

# Convert date to datetime format
combined_data['Date'] = pd.to_datetime(combined_data['Date'], format='%d-%m-%Y')

# Calculate the overall trend (best-fitting trend line) of the price data
X = np.arange(len(combined_data)).reshape(-1, 1)
y = combined_data['Close'].values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)
trend_line = model.predict(X)

# Determine the splitting points for the quarters
start_date = pd.to_datetime('2022-01-01')
quarter_size = pd.DateOffset(months=4)
split_points = [start_date + quarter_size, start_date + 2 * quarter_size]

# Plot the price graph and trend line for each quarter
plt.figure(figsize=(12, 18))

for i in range(3):
    start_date = split_points[i - 1] if i > 0 else pd.to_datetime('2022-01-01')
    end_date = split_points[i] if i < 2 else pd.to_datetime('2023-04-30')
    quarter_data = combined_data[(combined_data['Date'] >= start_date) & (combined_data['Date'] < end_date)]

    X_quarter = np.arange(len(quarter_data)).reshape(-1, 1)
    y_quarter = quarter_data['Close'].values.reshape(-1, 1)

    model.fit(X_quarter, y_quarter)
    trend_line_quarter = model.predict(X_quarter)

    # Subplot: Price plot as a step line
    plt.subplot(3, 1, i + 1)
    plt.plot(quarter_data['Date'], quarter_data['Close'], label=f'Quarter {i+1} Price', color='blue', drawstyle='steps-post')
    plt.plot(quarter_data['Date'], trend_line_quarter, label=f'Quarter {i+1} Trend Line', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Close price Trend - Quarter {i+1}')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
